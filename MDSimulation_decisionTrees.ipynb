```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
from sklearn.tree import DecisionTreeRegressor 
from sklearn import metrics
```

```python
filename = '/Users/vinitaboolchandani/surrogate_data/data4ML/surrogate/confined_ions/data/jcs_paper_data.dat'
```

```python
#data load
contents = open(filename).read().splitlines()
input_params=[]
output_density=[]
pos_val=[]
error_bars = []
for line in range(1, len(contents)):
            line_arr = contents[line].split(",")
            input_params.append(list(map(float, line_arr[0:5])))
            pos_val.append(list(map(float, line_arr[5:155])))
            output_density.append(list(map(float, line_arr[155:305])))
            error_bars.append(list(map(float, line_arr[305:455])))
train_input,test_input=input_params[0:2432],input_params[2432:]
train_gt,test_gt=output_density[0:2432],output_density[2432:]
```

```python
regressor = DecisionTreeRegressor(random_state = 0) 
regressor.fit(train_input, train_gt) 
```

```python
y_pred = regressor.predict(test_input)
print(len(y_pred))
```

```python
#print("Accuracy:",metrics.accuracy_score(test_gt, y_pred))
```

```python
for i in range(0,len(y_pred)):
    print('Mean Squared Error:', metrics.mean_squared_error(test_gt[i], y_pred[i]))
    #print("Accuracy:",metrics.accuracy_score(test_gt[i], y_pred[i]))
    #for j in range(0, len(y_pred[0])):
    #    del_y = test_gt[i][j] - y_pred[i][j]
```

```python
k=2432
theta = 0
for i in range(0,len(y_pred)):
    #print('Mean Squared Error:', metrics.mean_squared_error(test_gt[i], y_pred[i]))
    #print("Accuracy:",metrics.accuracy_score(test_gt[i], y_pred[i]))
    for j in range(0, len(y_pred[0])):
        x = abs(y_pred[i][j] - test_gt[i][j])
        if(x >= pos_val[k][j]):
            theta+= 0
        else:
            theta+= 1
    k+=1
A = 0.001 * theta 
print(A)
```
